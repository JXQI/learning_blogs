```python
import os
import re
import pandas as pd
```

```python
WORK_DIR = r'G:\01Download\temp\lunyu_20191012'
```

```python
CHAPTER_FILES = [
	["學而", "01 Xue Er.txt"],
	["為政", "02 Wei Zheng.txt"],
	["八佾", "03 Ba Yi.txt"],
	["里仁", "04 Li Ren.txt"],
	["公冶長", "05 Gong Ye Chang.txt"],
	["雍也", "06 Yong Ye.txt"],
	["述而", "07 Shu Er.txt"],
	["泰伯", "08 Tai Bo.txt"],
	["子罕", "09 Zi Han.txt"],
	["鄉黨", "10 Xiang Dang.txt"],
	["先進", "11 Xian Jin.txt"],
	["顏淵", "12 Yan Yuan.txt"],
	["子路", "13 Zi Lu.txt"],
	["憲問", "14 Xian Wen.txt"],
	["衛靈公", "15 Wei Ling Gong.txt"],
	["季氏", "16 Ji Shi.txt"],
	["陽貨", "17 Yang Huo.txt"],
	["微子", "18 Wei Zi.txt"],
	["子張", "19 Zi Zhang.txt"],
	["堯曰", "20 Yao Yue.txt"]
]
```

```python
for fname in os.listdir(WORK_DIR):
    print(fname)
```

```python
def extract_chapter_lines():
    text_list = []
    for chapter_idx, (chapter, fname) in enumerate(CHAPTER_FILES):
        print("processing file:{}".format(fname))
        fpath = os.path.join(WORK_DIR, fname)
        if not os.path.exists(fpath):
            print("can't find file {} for chapter:{}".format(fname, chapter))
            continue
        with open(fpath, encoding='UTF-8') as fi:
            for cnt, line in enumerate(fi):
                line = line.strip()
                if not line:
                    continue
                text_list.append([chapter_idx + 1, chapter, line])
                if cnt % 10 == 0:
                    print("processing cnt:{}".format(cnt))
    return text_list
```

```python
text_list = extract_chapter_lines()
```

```python
def char_count(line):
    if line is None:
        return 0
    return len(line)
```

```python
PUNC_REGEX = r'：|「|，|。|」|『|』|？|《|》|；|！|、'
def raw_char_count(line):
    raw_line = re.sub(PUNC_REGEX, '', line)
    return char_count(raw_line)
```

```python
def get_sentences(line):
    raw_sentences = re.split("：|，|。|；|？", line)
    sentences = []
    for s in raw_sentences:
        sentence = re.sub(PUNC_REGEX, '', s)
        if sentence:
            sentences.append(sentence)
    return sentences
```
